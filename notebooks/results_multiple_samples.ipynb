```python
def visualize_sample_density(samples, ground_truth_hist, ground_truth_fut, number_plots=5, number_samples=10):
    """
    Visualize the density of the samples generated by the model.

    :param samples: A list of predicted edge indices.
    :param ground_truth_hist: A list of actual history edge indices.
    :param ground_truth_fut: A list of actual future edge indices.
    """
    import matplotlib.pyplot as plt
    import networkx as nx
    import numpy as np
    
    G = nx.Graph()
    G.add_nodes_from(nodes)
    all_edges = {tuple(edge) for edge in edges}
    G.add_edges_from(all_edges)

    pos = nx.get_node_attributes(G, 'pos')

    for i in range(min(number_plots, len(samples))):
        plt.figure(figsize=(18, 8))            

        for plot_num, (title, edge_indices) in enumerate([
            ('Ground Truth History', ground_truth_hist[i][0]),
            ('Ground Truth Future', ground_truth_fut[i][0]),
            ('Predicted Future', samples[i])
        ]):
            plt.subplot(1, 3, plot_num + 1)
            plt.title(title)

            # Draw all edges as muted gray
            nx.draw_networkx_edges(G, pos, edgelist=all_edges, width=0.5, alpha=0.3, edge_color='gray')

            # Draw subgraph edges with specified color
            edge_color = 'gray' if plot_num == 0 else 'green' if plot_num == 1 else 'red'
            node_color = 'skyblue'
            if plot_num == 2:
                # if len(samples[i]) > 0 and not any(np.isnan(sample).any() or len(sample) == 0 for sample in samples[i]):
                if len(samples[i]) > 0:
                    edge_counts = {}
                    for sample in samples[i]:
                        for edge in sample:
                            edge = edge.item()
                            if edge in edge_counts:
                                edge_counts[edge] += 1
                            else:
                                edge_counts[edge] = 1
                    max_count = max(edge_counts.values(), default=1)
                    edge_widths = {edge: count / max_count for edge, count in edge_counts.items()}
                    unique_edges = list(set(edge_counts.keys()))
                    edge_list = {tuple(edges[idx]) for idx in unique_edges if idx < len(edges)}
                    nx.draw_networkx_nodes(G, pos, node_color=node_color, node_size=500)
                    nx.draw_networkx_edges(G, pos, edgelist=edge_list, edge_color='red', width=[edge_widths[edge] * 5 for edge in unique_edges], alpha=[edge_widths[edge] for edge in unique_edges])
                    # nx.draw_networkx_edges(G, pos, edgelist=all_edges, edge_color='red', width=edge_widths*5, alpha=edge_widths/np.max(edge_widths))
                    nx.draw_networkx_labels(G, pos, font_size=15)
            else:
                subgraph_edges = {tuple(edges[idx]) for idx in edge_indices if idx < len(edges)}
                nx.draw_networkx_nodes(G, pos, node_color=node_color, node_size=500)
                nx.draw_networkx_edges(G, pos, edgelist=subgraph_edges, width=3, alpha=1.0, edge_color=edge_color)
                nx.draw_networkx_labels(G, pos, font_size=15)
        plt.show()
        plt.close()  # Close the figure to free memory
```

```python
from collections import Counter

def get_n_most_common(samples, n):
    """
    Get the n most common numbers for each datapoint.

    :param samples: A list of lists of tensors where each sublist corresponds to multiple samples for one datapoint.
    :param n: The number of most common elements to return for each datapoint.
    :return: A list of lists containing the n most common numbers for each datapoint.
    """
    most_common_numbers = []
    
    for data_point_samples in samples:
        # Flatten the list of tensors into a single list of numbers
        all_numbers = torch.cat(data_point_samples).tolist()
        
        # Count the occurrences of each number
        counter = Counter(all_numbers)
        
        # Find the n most common numbers
        most_common = counter.most_common(n)
        
        # Extract just the numbers (without counts)
        most_common_numbers.append([item[0] for item in most_common])
    
    return most_common_numbers
```

```python
def calculate_fut_ratio(sample_list, ground_truth_fut):
    count_1 = 0
    count_2 = 0
    total = len(sample_list)
    
    for i, sample in enumerate(sample_list):
        edges_count = sum(1 for edge in ground_truth_fut[i][0] if edge in sample)
        if edges_count >= 1:
            count_1 += 1
        if edges_count >= 2:
            count_2 += 1
    
    ratio_1 = count_1 / total
    ratio_2 = count_2 / total
    return ratio_1, ratio_2

def calculate_hist_ratio(sample_list, ground_truth_hist):
    count_1 = 0
    count_2 = 0
    count_3 = 0
    count_4 = 0
    count_5 = 0
    total = len(sample_list)
    
    for i, sample in enumerate(sample_list):
        edges_count = sum(1 for edge in ground_truth_hist[i][0] if edge in sample)
        if edges_count >= 1:
            count_1 += 1
        if edges_count >= 2:
            count_2 += 1
        if edges_count >= 3:
            count_3 += 1
        if edges_count >= 4:
            count_4 += 1
        if edges_count >= 5:
            count_5 += 1
    
    ratio_1 = count_1 / total
    ratio_2 = count_2 / total
    ratio_3 = count_3 / total
    ratio_4 = count_4 / total
    ratio_5 = count_5 / total
    return ratio_1, ratio_2, ratio_3, ratio_4, ratio_5
```
